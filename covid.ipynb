```python
from folium import plugins

# let's start again with a clean copy of the world map
coronamap2 = folium.Map( tiles='Stamen Toner')

# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(coronamap2)

# loop through the dataframe and add each data point to the mark cluster
for lat,lng,total_cases,total_death,total_recovered,serious in zip(df["Latitude"], df["Longitude"],df["total_cases"],df["total_death"],df["total_recovered"],df["serious"]):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=("total_cases"+"-"+str(total_cases)+'<br>'+"total_death%"+"-"+str(total_death)+'<br>'+"total_recovered"+"-"+str(total_recovered)+'<br>'+"serious"+"-"+str(serious)),
    ).add_to(incidents)
coronamap2
```
